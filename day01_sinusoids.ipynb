```python
import numpy as np
import matplotlib.pyplot as plt
```

Plot:

$y=A\cos(\omega t - \phi)$

```python
t = np.linspace(-2*np.pi, 2*np.pi,101)
y = -np.sin((np.pi/2)*t)
fig = plt.figure(figsize=(10, 5), facecolor='white')
ax = fig.add_subplot(111)
ax.axhline(color="black")
ax.plot(t,2*y)
ax.plot(t, -2*np.sin(t), color="red")
plt.show()
```

```python
A = 2.0
omega = 1.5
phi = 0
t = np.linspace(-np.pi, np.pi,201)
new_y = A*np.cos(omega*t-phi)
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111) # Adds a single subplot

ax.plot(t,new_y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Plot in Python')

plt.show() 
```
