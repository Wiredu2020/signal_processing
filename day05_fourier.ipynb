```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1001
T0 = 2
w0 = 2*np.pi/T0

t = np.linspace(-T0/2, T0/2, N)
X = 2*t/T0
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(111)
ax.axhline(color = "black")
ax.axvline(color = "black")
ax.plot(t,X)
plt.show()
```

```python
kmax = 1000
t = np.linspace(-T0*2, T0*2, N)
x_f = np.zeros(N, dtype=float)
for k in range(1, kmax):
    x_f += (-2/np.pi)* (-1)**k / k * np.sin(k*w0*t)
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(111)
ax.axhline(color = "black")
ax.axvline(color = "black")
ax.plot(t,x_f)
plt.show()
```
